```python
from pyspark.sql import SparkSession

# 1. Configurar la sesión para descargar el conector de Cassandra automáticamente
spark = SparkSession.builder \
    .appName("Validacion_Fase_1_4") \
    .config("spark.jars.packages", "com.datastax.spark:spark-cassandra-connector_2.12:3.5.0") \
    .config("spark.cassandra.connection.host", "cassandra_db") \
    .getOrCreate()


data = [("Cassandra", "Enlazada"), ("PySpark", "Operativo"), ("ClickHouse", "Listo")]
df_validacion = spark.createDataFrame(data, ["Componente", "Estado"])

# 2. Mostrar el resultado 
print("✅ HOLA MUNDO: ENTORNO BIG DATA CONFIGURADO")
df_validacion.show()

# 3. Imprimir la confirmación del paquete cargado
print("Paquete detectado:", spark.sparkContext.getConf().get("spark.jars.packages"))
```
