```python
import sys
import types

# Simulaci√≥n del m√≥dulo 'asyncore' requerido por cassandra-driver en versiones recientes de Python
mock_as = types.ModuleType('asyncore')
class MockDispatcher:
    def __init__(self, *args, **kwargs): pass
    def close(self): pass
    def add_channel(self, map=None): pass
    def del_channel(self, map=None): pass
mock_as.dispatcher = MockDispatcher
mock_as.socket_map = {}
sys.modules['asyncore'] = mock_as

print("‚úÖ Entorno de compatibilidad configurado.")
```

```python
from cassandra.cluster import Cluster
import sys

if 'asyncore' not in sys.modules:
    import types
    sys.modules['asyncore'] = types.ModuleType('asyncore')

print("üèóÔ∏è Configurando Keyspace y Tablas en Cassandra...")

# Conexi√≥n directa por red al contenedor de Cassandra
# 'cassandra_db' es el nombre del servicio en docker-compose
cluster = Cluster(['cassandra_db']) 
session = cluster.connect()

# Ejecuci√≥n de los comandos CQL directamente
session.execute("CREATE KEYSPACE IF NOT EXISTS proyecto_bigdata WITH replication = {'class': 'SimpleStrategy', 'replication_factor': 1};")
session.execute("CREATE TABLE IF NOT EXISTS proyecto_bigdata.ventas_crudas (id int PRIMARY KEY, categoria text, precio float, fecha timestamp);")

print("‚úÖ Infraestructura de Cassandra lista.")
cluster.shutdown()
```

```python
import clickhouse_connect
from cassandra.cluster import Cluster

print("üèóÔ∏è CONFIGURACI√ìN DE ESQUEMAS (FASE 1.5)")

# --- 1. CONFIGURACI√ìN EN CASSANDRA ---
try:
    cluster = Cluster(['cassandra_db'])
    session = cluster.connect()
    session.execute("CREATE KEYSPACE IF NOT EXISTS proyecto_bigdata WITH replication = {'class': 'SimpleStrategy', 'replication_factor': 1};")
    session.execute("""
        CREATE TABLE IF NOT EXISTS proyecto_bigdata.ventas_crudas (
            id int PRIMARY KEY, 
            categoria text, 
            precio float, 
            fecha timestamp
        );
    """)
    print("‚úÖ Cassandra: Tabla 'ventas_crudas' lista.")
    cluster.shutdown()
except Exception as e:
    print(f"‚ùå Error en Cassandra: {e}")

# --- 2. CONFIGURACI√ìN EN CLICKHOUSE ---
try:
    # Usamos 'clickhouse_dw' que es el host interno de Docker
    client = clickhouse_connect.get_client(host='clickhouse_dw', port=8123, username='default', password='')
    
    # Crear la tabla ventas_resumen (Esquema de destino)
    client.command('DROP TABLE IF EXISTS ventas_resumen') # Opcional: para limpiar antes de configurar
    client.command('''
        CREATE TABLE ventas_resumen (
            categoria String, 
            total_ventas Float64, 
            cantidad_transacciones UInt64
        ) ENGINE = MergeTree()
        ORDER BY categoria
    ''')
    print("‚úÖ ClickHouse: Tabla 'ventas_resumen' lista.")
except Exception as e:
    print(f"‚ùå Error en ClickHouse: {e}")
```

```python
import pandas as pd
import clickhouse_connect
from cassandra.cluster import Cluster
from IPython.display import display, HTML

print("üîç VERIFICANDO INFRAESTRUCTURA DE DATOS...")

try:
    # --- 1. VERIFICACI√ìN DE CASSANDRA ---
    cluster = Cluster(['cassandra_db'])
    session = cluster.connect('proyecto_bigdata')
    
    # Al usar '*', la consulta funciona con cualquier esquema que hayas definido
    res_cas = session.execute("SELECT * FROM ventas_crudas LIMIT 1;")
    
    display(HTML("<h3 style='color: #2e7d32;'>üü¢ CASSANDRA: Conectividad Exitosa</h3>"))
    print("La tabla 'ventas_crudas' est√° disponible y lista para la ingesta de datos.")
    cluster.shutdown()

    # --- 2. VERIFICACI√ìN DE CLICKHOUSE ---
    client = clickhouse_connect.get_client(host='clickhouse_dw', port=8123, username='default', password='')
    
    # Obtenemos la estructura t√©cnica para validar el esquema
    res_ch = client.query('DESCRIBE TABLE ventas_resumen')
    df_ch_schema = pd.DataFrame(res_ch.result_rows, columns=['Columna', 'Tipo', 'Diferenciador', 'Key', 'Default', 'Extra', 'Codex'])

    display(HTML("<h3 style='color: #1565c0;'>üîµ CLICKHOUSE: Esquema de Destino</h3>"))
    display(df_ch_schema[['Columna', 'Tipo']])

    print("\n‚úÖ Los servicios est√°n sincronizados y los esquemas configurados correctamente.")

except Exception as e:
    print(f"‚ùå Error de conectividad: {e}")
```
